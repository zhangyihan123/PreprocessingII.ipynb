```python
import pandas as pd
# Read the dataset
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/postoperative-patient-data/post-operative.data"
names = ['L-CORE', 'L-SURF', 'L-O2', 'L-BP', 'SURF-STBL', 'CORE-STBL', 'BP-STBL', 'COMFORT', 'decision ADM-DECS']
df = pd.read_csv(url, names = names, comment='#')
print (df)
```

# Preprocessing

```python
df.columns = ['L-CORE', 'L-SURF', 'L-O2', 'L-BP', 'SURF-STBL', 'CORE-STBL', 'BP-STBL', 'COMFORT', 'decision ADM-DECS']
df
```

We need to transfer the data into a clear form.

We need to eliminate the rows with missing data so that we may get accurate learning of data.

Then we may analyse the data separately.

## L-CORE

```python
aa = df.loc[:,["L-CORE", "COMFORT"]]
ab = pd.DataFrame(aa)
ab.columns = ['size', 'comfort']
size_mapping = {'high': 3,'mid': 2,'low': 1}
ab['size'] = ab['size'].map(size_mapping)
print (ab)
ab.boxplot(column='size', by='comfort')
```

## L-SURF

```python
ba = df.loc[:,["L-SURF", "COMFORT"]]
bb = pd.DataFrame(ba)
bb.columns = ['surf', 'comfort']
size_mapping = {'high': 3,'mid': 2,'low': 1}
bb['surf'] = bb['surf'].map(size_mapping)
print (bb)
bb.boxplot(column='surf', by='comfort')
```

## L-O2

```python
ca = df.loc[:,["L-O2", "COMFORT"]]
cb = pd.DataFrame(ca)
cb.columns = ['O2', 'comfort']
size_mapping = {'excellent': 2,'good': 1}
cb['O2'] = cb['O2'].map(size_mapping)
print (cb)
cb.boxplot(column='O2', by='comfort')
```

## L-BP

```python
da = df.loc[:,["L-BP", "COMFORT"]]
db = pd.DataFrame(da)
db.columns = ['BP', 'comfort']
size_mapping = {'high': 3,'mid': 2,'low': 1}
db['BP'] = db['BP'].map(size_mapping)
print (db)
db.boxplot(column='BP', by='comfort')
```

## SURF-STBL

```python
ea = df.loc[:,["SURF-STBL", "COMFORT"]]
eb = pd.DataFrame(ea)
eb.columns = ['STBL', 'comfort']
size_mapping = {'unstable': 2,'stable': 1}
eb['STBL'] = eb['STBL'].map(size_mapping)
print (eb)
eb.boxplot(column='STBL', by='comfort')
```

## CORE-STBL

```python
fa = df.loc[:,["CORE-STBL", "COMFORT"]]
fb = pd.DataFrame(fa)
fb.columns = ['CORE', 'comfort']
size_mapping = {'unstable': 2,'stable': 1}
fb['CORE'] = fb['CORE'].map(size_mapping)
print (fb)
fb.boxplot(column='CORE', by='comfort')
```

## BP-STBL

```python
ga = df.loc[:,["CORE-STBL", "COMFORT"]]
gb = pd.DataFrame(ga)
gb.columns = ['BS', 'comfort']
size_mapping = {'mod-stable': 3, 'unstable': 2,'stable': 1}
gb['BS'] = gb['BS'].map(size_mapping)
print (gb)
gb.boxplot(column='BS', by='comfort')
```

## decision ADM-DECS

```python
ha = df.loc[:,["decision ADM-DECS", "COMFORT"]]
hb = pd.DataFrame(ha)
hb.columns = ['dec', 'comfort']
size_mapping = {'A': 3, 'I': 2,'S': 1}
hb['dec'] = hb['dec'].map(size_mapping)
print (hb)
hb.boxplot(column='dec', by='comfort')
```
